```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the data
df1 = pd.read_csv("CycPeptMPDB_Peptide_Assay_PAMPA.csv", low_memory=False)
df2 = pd.read_csv("mordred_colab.csv", low_memory = False)
```

```python
df1
```

```python
def convert_columns_to_float(df):
    for col in df.columns:
        try:
            df[col] = df[col].astype(float)
        except ValueError:
            df.drop(columns=[col], inplace=True)
    return df
```

```python
df1 = convert_columns_to_float(df1)
```

```python
df2 = convert_columns_to_float(df2)

df2['Permeability'] = df1['Permeability']
```

```python
def classify_permeability(value):
    if value >= -5:
        return 'good'
    elif -7 < value < -5:
        return 'moderate/low'
    else:
        return 'impermeable'

df1['Permeability'] = df1['Permeability'].apply(classify_permeability)

print(df1['Permeability'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.set(style="whitegrid", palette = "muted")
sns.scatterplot(
    x='PC1', 
    y='PC2', 
    hue='Permeability', 
    linewidth=0.5,
    data=df1,
)

plt.title('Principal component analysis - 2D features')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Permeability')
plt.savefig("PC_2D_features.jpg")
```

```python
df2 = df2[df2.Permeability != -10]
```

```python
def classify_permeability(value):
    if value >= -5:
        return 'good'
    elif -7 < value < -5:
        return 'moderate/low'
    else:
        return 'impermeable'

df2['Permeability'] = df2['Permeability'].apply(classify_permeability)

# Display the distribution of the new classes
print(df2['Permeability'].value_counts())
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

temp = df2['Permeability']
df2 = df2.drop(['Permeability'], axis = 1)

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df2)

pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_data)

pc_df = pd.DataFrame(data=principal_components, columns=['PC3', 'PC4'])
```

```python
pc_df['Permeability']= temp
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid", palette="muted")

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='PC3', 
    y='PC4', 
    hue='Permeability', 
    data=pc_df,
    linewidth=0.5,
)

plt.title('Principal component analysis - 3D features')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Permeability')
plt.savefig("PC_3D_features.jpg")
```
